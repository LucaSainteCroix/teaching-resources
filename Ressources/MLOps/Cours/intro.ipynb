## What is MLOps or "Machine Learning Operations"?

--> A set of practices that automate and simplify machine learning (ML) workflows and deployments.

## Why MLOps? Example with a notebook

- Experimenting with models - experiment tracking and model management (MLFlow)
    - track training & eval dataset
    - track metrics for each training
    - save model + scalers, vectorizers...
- Retraining models - ML pipelines for automation and reproducability
    - Prefect
    - Kubeflow
- Serving models (deployment)
- Monitoring models
- Automate Everything
- Best Practices

## MLOps Maturity Model

[Source](https://learn.microsoft.com/en-us/azure/architecture/ai-ml/guide/mlops-maturity-model)

### Level 0 : No MLOps

Everything is done manually and every team works separately.

### Level 1 : DevOps but no MLOps

Basically when you go directly from Proof Of Concept to Production


- Releases are automated
- Unit & Integration tests
- CI/CD
- Ops Metrics (How many requests per second etc.)


- No Experiment Tracking
- No reproducibility
- Teams (Data Science, Data Engineering...) work separately
- Data pipeline might or might not gather data automatically 

### Level 2 : Automated Training

Necessary when there are more than 1 ML use case in the company


- Training Pipeline
- Experiment Tracking
- Model Registry
- Low Friction Deployment
- DS works with Engineering Department


- No Automated Deployment
- No ML metrics monitoring

### Level 3 : Automated Deployment

Necessary when there are more ML use cases or more mature/important/scaled ML use cases 


- Easy to Deploy Model
    - ML Platform to deploy, store and make model available
    - A/B Tests possible
    - Model Monitoring


- No automatic retraining

### Level 4 : Automated Training, Retraining & Deployment

Full MLOps implementation


- Retraining triggered automatically based on production metrics
- More advanced monitoring
- Just in general advanced awareness of best MLOps practices by the teams

--> No humans needed in the loop (ideally)
